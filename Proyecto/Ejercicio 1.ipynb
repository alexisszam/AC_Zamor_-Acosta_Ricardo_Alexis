```python
import numpy as np                          #biblioteca de trazado de Python
import matplotlib.pyplot as plt             #interfaz para representación gráfica
import random                               #funciones "random"
```

```python
A = [0, 10, 5]                   #Definimos los vertices de nuestro tránuglo equilatero pero sin graficarlos
B = [0, 0, 20]
```

```python
L1 = np.random.rand()            #Definimos nuestro punto aleatorio dentro del triángulo
L2 = (1 - L1)
```

```python
plt.plot(L1, L2, 'mo')           #Graficamos nuestro punto

L1 = [L1]
L2 = [L2]
```

```python
x = 0                            
n = 10                           #10 iteraciones
l = [0, 1, 2]                    #escoja un valor de l aleatoriamente

while x < n:
    i = random.choice(l)
    L1.extend([(L1[-1] + A[i]) / 2])   #fórmula para la obtención de puntomedio
    L2.extend([(L2[-1] + B[i]) / 2])
    x += 1
    
plt.plot(L1, L2, 'mo')           #Grafique 
```

```python
def Puntomedio(n):                    #Una vez que funcionó lo anterior generamos una función para que se haga para n iteraciones
    A = [0, 10, 5]
    B = [0, 0, 20]
    L1 = np.random.rand()
    L2 = (1 - L1)                      #Definimos las variiables que usaremos
    L1 = [L1]
    L2 = [L2]
    x = 0
    l = [0, 1, 2]
    
    while x < n:                      #Ciclo while
        i = random.choice(l)
        L1.extend([(L1[-1] + A[i]) / 2])    #Fórmula
        L2.extend([(L2[-1] + B[i]) / 2])  
        x += 1
        
    plt.plot(L1, L2, 'ko')             #Grafica
```

```python
Puntomedio(10)
```

```python
Puntomedio(1000)
```

```python
Puntomedio(10000)
```
